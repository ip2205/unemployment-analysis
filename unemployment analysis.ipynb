### Q1. What is average estimated unemployment rate state wise?

```python
unemployment_df
# df.groupby([df['ord_date'].dt.year, df['ord_date'].dt.month])
```

We use groupby() function to categorize Regions state wise. It will group all rows state wise. We can get ouput according to our requirement. Suppose we want average of unemployment rate than we can use.mean() so that we get mean of rate in region.

If we sum() instead of mean we will get sum of all entities in every column.

```python
group_by_region_df= unemployment_df.groupby(by='Region').mean().sort_values(by=' Estimated Unemployment Rate (%)',ascending=False)
group_by_region_df
```

### Q2. How to get overall unemployment rate of specific given state?
Here for example we take state Maharashtra.

```python
unemp_state_df= unemployment_df[unemployment_df.Region=='Maharashtra']
unemp_state_df
```

Above we got all Maharashtra's information by frequency of month

Now in below cell we use .mean() to caslculate mean for specific column.

We want average unemployment rate so we use mean on unemployment rate column.

```python
unemp_state_df[' Estimated Unemployment Rate (%)'].mean()
```

We get average unemployment rate in Maharashtra is approx 7.56 %.

### Q3. Which state has highest unemployment rate overall?

We use sort_values function to sort values. By default it sorts elements in Ascending order, so we have to pass an argument **ascending=False** along with **by argument** where we pass sorting has to be taken according to which column.

```python
group_by_region_df.sort_values(by=' Estimated Unemployment Rate (%)',ascending=False)
```

From above table we can observe that highest unemployment rate is in Tripura with 28.350357 %. And lowest in Meghalaya.

### Q4. What is unemployment average rate of specified state in specific year?

```python

maharashtra_2020[' Estimated Unemployment Rate (%)'].mean()
```

We calculate average unemployment in Maharashtra using mean().

### Chart 1. Using Line Chart to see increase in rise of unemployment

```python
years=list(set(list(unemployment_df[' Date'].dt.year)))
gbyyear_df=unemployment_df.groupby(by='year')
# m=list(unemployment_df[' Estimated Unemployment Rate (%)'])
# m
mean_uemployment_rate_overall =list(gbyyear_df[' Estimated Unemployment Rate (%)'].mean())
```

```python
mean_uemployment_rate_overall
# years
```

```python
sns.set_style("darkgrid")
plt.plot(years, mean_uemployment_rate_overall, 's-b')
# plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('mean_uemployment_rate_overall(%)')

plt.title("Unemployment rate by Year")
# plt.legend(['Apples', 'Oranges']);
```

From the above plot it is clear that unemployment rate in increased as compared to earlier years.
We don't have data for year 1970 . 
But if we compare year 2019 and 2020 only we can observe huge rise in unemployment from  9.40% to 15.10%.
